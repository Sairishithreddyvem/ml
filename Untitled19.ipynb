```python
import pandas as pd
```

```python
df=pd.read_csv('/Salary_dataset.csv')
```

```python
df.head()
```

```python
x=df.YearsExperience
y=df.Salary
```

```python
import numpy as np
```

```python
x=np.array(x).reshape(-1,1)
y=np.array(y).reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(
    x,y,
    test_size=0.2,
    random_state=0)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print("Actual salary:",y_test[:5])
print("Predicted salary:",y_pred[:5])
```

```python
#calculate mean squared erroer
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(y_test,y_pred)
print("mean squared error:",mse)
```

```python
#calculate RMSE
```

```python
rmse=np.sqrt(mse)
print("root mean squared error:",rmse)
```

```python
#calculate R^2 value
```

```python
from sklearn.metrics import r2_score
```

```python
r2=r2_score(y_test,y_pred)
print("R^2 value:",r2)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x_test, y_test, color='red')
plt.plot(x_test, y_pred, color='blue')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
plt.scatter(x_test, y_test, label="actual salary")
plt.scatter(x_test, y_pred, label="predict salary")
plt.title('actual vs predicted salary comparision')
plt.xlabel('actual salary')
plt.ylabel('salary value')
plt.legend()
plt.show()
```

```python
slope=model.coef_
intercept=model.intercept_
print("Slope (m):",slope)
print("Intercept (c):",intercept)
```

```python
experience=[[5]]
```

```python
predict_salary=model.predict(experience)
print("Predicted salary for 5 year experience:",predict_salary)
```

```python
salary_manually=slope*5+intercept
print("Predicted salary for 5 year experience (manually):",salary_manually)
```
